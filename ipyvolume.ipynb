# Ipyvolume

```python
import ipyvolume as ipv
import ipywidgets as widgets
import numpy as np
```

```python
from astropy.io import fits
hdulist = fits.open('ngc6946.fits')
cube = hdulist[0].data
#print(np.nanmin(cube), np.nanmax(cube))
cube[np.isnan(cube)] = 0
# import scipy.ndimage
# cube = scipy.ndimage.gaussian_filter(cube, 2.5)
```

```python
ipv.figure()
ipv.volshow(cube, level=[0.3, 0.11, 0.19], 
                  opacity=[0.2, 0.1, 0.06], level_width=0.05, data_min=0.03, data_max=0.15)
ipv.show()
```

```python
ipv.figure()
N = 100
x, y, z = np.random.random((3, N))
ipv.figure()
scatter = ipv.scatter(x, y, z, color='orange', marker='sphere')
ipv.show()
```

```python
scatter.x = 0.5
```

```python
color_picker = widgets.ColorPicker()
widgets.jslink((scatter, 'color'), (color_picker, 'value'))
color_picker
```

```python
size_slider = widgets.FloatSlider(min=0.1, max=5)
widgets.jslink((scatter, 'size'), (size_slider, 'value'))
size_slider
```

```python
ipv.figure()
N = 100
M = 10
x, y, z = np.random.random((3, M, N))
vx, vy, vz = np.random.random((3, M, N))*2 - 1
ipv.figure()
quiver = ipv.quiver(x, y, z, vx, vy, vz, color=np.random.random((M,N,3)))
ipv.animation_control(quiver, interval=1000)
ipv.show()
```

```python
ipv.figure()
ipv.style.use('dark')
quiver = ipv.quiver(*ipv.datasets.animated_stream.fetch().data[:,::,::], size=5)
ipv.animation_control(quiver, interval=200)
ipv.show()
```
